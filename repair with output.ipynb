```python
def run_simulation(sddp, mode):
    # Create an in-memory text stream
    memory_file = io.StringIO()

    if mode == 'Fidelity':
      # Redirect stdout to the in-memory stream
      with redirect_stdout(memory_file):
          medrasrepair.repairSimulation(sddpath,'Fidelity', 24)

      # Reset the stream position to the beginning
      memory_file.seek(0)

      # Read the data from the in-memory stream
      repair_data = [line.split() for line in memory_file]# if line[0].isdigit()]

      # Close the in-memory stream
      memory_file.close()

      return repair_data

    elif mode == 'Spectrum':
      # Redirect stdout to the in-memory stream
      with redirect_stdout(memory_file):
          medrasrepair.repairSimulation(sddp, 'Spectrum')
          #repair_data = medrasrepair.repairSimulation(sddp, 'Spectrum')

      # Reset the stream position to the beginning
      memory_file.seek(0)

      # Read the data from the in-memory stream if the line isnt empty
      #repair_data = [line for line in memory_file]
      repair_data = [line.split() for line in memory_file]
      #repair_data = [line.split() for line in memory_file if line[0].isdigit()]

      # Close the in-memory stream
      memory_file.close()

      return repair_data

    elif mode == "postRepairDNA":
          # Redirect stdout to the in-memory stream
      with redirect_stdout(memory_file):
          medrasrepair.repairSimulation(sddp, 'postRepairDNA')
          #repair_data = medrasrepair.repairSimulation(sddp, 'Spectrum')

      # Reset the stream position to the beginning
      memory_file.seek(0)

      # Read the data from the in-memory stream if the line isnt empty
      #repair_data = [line for line in memory_file]
      repair_data = [line.split() for line in memory_file]
      #repair_data = [line.split() for line in memory_file if line[0].isdigit()]

      # Close the in-memory stream
      memory_file.close()

      return repair_data
    else:
      print('Invalid mode')
      return None

def getpath(sddp):
  for f in os.listdir(sddp):
    if f.endswith(".txt"):
      return f.split('.')[0]
```
